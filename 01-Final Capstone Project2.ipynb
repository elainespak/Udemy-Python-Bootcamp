```python
# Fetch Current Weather
# Optional: Try locating the user automatically.

# Import necessary packages

import geocoder
from bs4 import BeautifulSoup
import requests
from uszipcode import ZipcodeSearchEngine
```

```python
# Get user's zip code
while True:
    zipcode = str(input('Type in your 5-digit zip code: '))
    #print('Invalid, please try again.')
    if len(zipcode) == 5 and zipcode.isdigit() == True:
        print(f'Your zip code is {zipcode}.')
        break
    else:
        print('Invalid, zip code has to be a number and 5-digit.')
```

```python
# uszipcode package finds the city and state of the given zipcode
search = ZipcodeSearchEngine()
info = search.by_zipcode(zipcode)

city = info['City'].lower().replace(' ', '')
state = info['State'].lower()
```

```python
url = 'https://www.wunderground.com/weather/us' + '/' + state + '/' + city + '/' + str(zipcode)
page = requests.get(url, timeout = 5)
content = BeautifulSoup(page.content, 'html.parser')

feels_like = content.find(class_='feels-like')
weather_quickie = content.find(class_='weather-quickie')
timestamp = content.find(class_='timestamp')
condition_icon = content.find(class_='condition-icon small-6 medium-12 columns').find('p')
condition_wind = content.find(class_='condition-wind small-6 medium-12 columns')

weather = list(feels_like)[0] + list(feels_like)[1].get_text()
update = list(timestamp)[4].get_text() + 'at' + list(timestamp)[1].get_text()
tomorrow = list(weather_quickie)[0] + list(weather_quickie)[1].get_text() + list(weather_quickie)[2]
prec = condition_icon.get_text()
wind = condition_wind.find(class_='wind-speed').find('strong').get_text() + ' ' + condition_wind.find('p').get_text()
```

```python
print(f'Weather Information in {zipcode}: \n {weather} \n {update} \n {tomorrow} \n {prec} \n {wind}')
```
