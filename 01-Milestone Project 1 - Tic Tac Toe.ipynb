```python
def symbol():
    global p1_symbol
    global p2_symbol
    while True:
        p1_symbol = input("Player 1, please choose between 'X' and 'O': ")
        if p1_symbol == 'X':
            p1_symbol = 'X'
            p2_symbol = 'O'
            print("Player 1's symbol is set to 'X'; Player 2's symbol to 'O'")
            break
        elif p1_symbol == 'O':
            p1_symbol = 'O'
            p2_symbol = 'X'
            print("Player 1's symbol is set to 'O'; Player 2's symbol to 'X'")
            break
        else:
            print("Please try again.")

def display_board(board):
    hor = "-----"
    print(hor)
    print(f"{board[0]}|{board[1]}|{board[2]}")
    print(hor)
    print(f"{board[3]}|{board[4]}|{board[5]}")
    print(hor)
    print(f"{board[6]}|{board[7]}|{board[8]}")
    pass
```

```python
def player_input(board, player):
    global position
    while True:
        position = int(input(f"{player}, choose a number: "))
        if position < 0 or position > 8:
            print(f"Invalid, {player}, please try again.")
        elif board[position] == 'X' or board[position] == 'O':
            print(f"Invalid, {player}, please try again.")
        else:
            break

def place_marker(board, marker, position):
    board[position] = marker
    display_board(board)
    pass

def check_win(board):
    if board[0] == board[1] == board[2]:
        return True
    if board[3] == board[4] == board[5]:
        return True
    if board[6] == board[7] == board[8]:
        return True
    if board[0] == board[3] == board[6]:
        return True
    if board[1] == board[4] == board[7]:
        return True
    if board[2] == board[5] == board[8]:
        return True
    if board[0] == board[4] == board[8]:
        return True
    if board[2] == board[4] == board[6]:
        return True
```

```python
def game():
    # Set the game up here
    board = list(range(0,10))
    turns = 9
    
    #choose_first() -- ignore for now
    symbol()
    display_board(board)

    while turns > 0:
        #Player1's turn
        if turns % 2 == 1:
            player1 = "Player 1"
            player_input(board, player1)
            place_marker(board, p1_symbol, position)
            if check_win(board):
                print(f"Well done, {player1}, you won!")
                break
            else:
                pass
        # Player2's turn
        else:
            player2 = "Player 2"
            player_input(board, player2)
            place_marker(board, p2_symbol, position)
            if check_win(board):
                print(f"Well done, {player2}, you won!")
                break
            else:
                pass
        
        turns -= 1
    #print("There is no winner!")
```

```python
game()
```
